```python
#indepentdent variables
features = ["national_inv","lead_time","sales_3_month","sales_6_month","forcast_3_month","forcast_6_month"]

```

```python
X=df2[features]
Y=df2["went_on_backorder"]
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(X)
X= scaler.transform(X)
X=pd.Dataframe(X,columns=features)
```

```python
X1=X.copy()
Y1=Y.copy()
```

```python
#splittng the data in training and testing
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.1,stratify=Y,random_state=100)
```

```python
#model building
```

```python
from sklearn.matrics import roc_auc_score,confusion_matrix,accuracy_score,classification_report
```

```python
#decision tree
from sklearn.tree import DecisionTreeClassifier
```

```python
dt=DecisionTreeClassifier()
dt.fit(X_train,Y_train)
```

```python
Y_pred_dt=dt.predict(X_test)
```

```python
#model performance
print("".dt.score(X_train,Y_train))

print(accuracy_score(Y_test,Y_pred_dt))
print(confusion_Matrix(Y_test,Y_pred_dt))
print(classification_report(Y_test,Y_pred_dt))
```
